```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
#class LTF
class LIF:
    
    def __init__(self):  
        
        self.theta=30               ## Threshold value
        self.dt=0.1               ## step
        
    def Response(self,a,b,v_res,u_res,Input):
        
        self.a=a               
        self.b=b
        self.v_res=v_res             
        self.u_res=u_res
        T=np.shape(Input)[0]
        v=np.zeros([T])
        u=np.zeros([T])
        Spike=np.zeros([T])
        
        v[0]=self.v_res
        u[0]=self.v_res*self.b    

        for t in range(1,T):

            temp_v=(0.04*(v[t-1]**2)+5*v[t-1]+140-u[t-1]+Input[t-1])
            v[t]=v[t-1]+temp_v*self.dt    
            temp_u=(a*(b*v[t-1]-u[t-1]))
            u[t]=u[t-1]+temp_u*self.dt
            
            ## If the activity is grater then a threshold, the neuron fires and the the variable spike=1, otherwise 0
            spike=np.float32(v[t]>=self.theta)
            Spike[t]=np.copy(spike)
            
            ## Reset 
            v[t]=self.v_res*spike+v[t]*(1-spike)
            u[t]=(1-spike)*u[t]+(u[t]+self.u_res)*spike
            
        return v, Spike
            
```

```python
lif=LIF()

T=3000
Input=np.concatenate([np.zeros([300]),np.ones([T-300])*10])        
                                                                       

a=0.1
b=0.2
v_res=-65
u_res=8                   
    
x,Spike=lif.Response(a,b,v_res,u_res,Input)       ## Computation of the neuron response

# Creation of the figure

plt.rcParams.update({'font.size': 35})

fig, (ax1,ax2) = plt.subplots(2,1,figsize=(10, 8))
plt.rcParams['figure.constrained_layout.use'] = True


# Plot of the activity of the neuron across time
ax1.plot(np.arange(0,T)*lif.dt,x,color='black')

# Plot of the input
ax2.plot(np.arange(0,T)*lif.dt,Input,color='black')


# Set labels
ax1.set_xlabel('t')
ax1.set_ylabel('v(t)')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)


ax2.set_xlabel('t')
ax2.set_ylabel('v(t) (Input)')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

plt.suptitle('Fast Spiking')
```
